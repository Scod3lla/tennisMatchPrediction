```python
# Swap tra due valori
def swap (a,b):
    c=a
    a=b
    b=c
    return a,b
```

```python
import numpy as np
def randomSwap(df, randomNumber):
    copy = df.copy()
    swapCol=["Player1", "Player2"]
    for i in copy.columns:
        if "P1" in i or "P2" in i:
            swapCol.append(i)

    maxRow = copy.shape[0]
    print("Colonne")
    print(swapCol)
    for i in range(0,randomNumber):
        pos = np.random.randint(0,maxRow)
        for i in range(0, int(len(swapCol)/2)):
            index=i*2
            aa = swapCol[index]
            bb = swapCol[index+1]
            print("pre",copy.loc[pos,aa],copy.loc[pos,bb])
            copy.loc[pos,aa], copy.loc[pos,bb] = swap(copy.loc[pos,aa],copy.loc[pos,bb])
            print("post",copy.loc[pos,aa],copy.loc[pos,bb])
            
        if copy.loc[pos,"target"] == 0:
            copy.loc[pos,"target"] = 1
        else:
            copy.loc[pos,"target"] = 0
        
    return copy
```
